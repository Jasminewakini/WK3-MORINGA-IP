# Data Cleaning

```python
# To start of the data cleaning, reducing the number of tables by joining the 3 Telcom tables into one
#
df_tel = pd.concat([df_1,df_2,df_3], axis=0)
df_tel
```

```python
# Looking at the table's info
df_tel.info()
```

```python
# From the above info we can see that the 'Date_Time' column has a string data type..
# We can correct this by changing it to he correct data type datetime
#
df_tel['Date_Time'] = pd.to_datetime(df_tel['Date_Time'])
df_tel

# To check if Data type changed, we look at the info
df_tel.info()
```

```python
# To view our current table
df_tel

# From the table we have some columsns that are not required for our analysis, therefore we can drop them

df_tel2 = df_tel.drop(['Country_A', 'Country_B'], axis=1)
df_tel2
```

```python
# From our data exploration section we had identified that we had duplicates in our tables, because we had joined our Telcom tables into one,
# dropping the duplicates here will be much easier than dropping them for each individual table

df_tel3 = df_tel2.drop_duplicates()
df_tel3
```

```python
# With a clean Telcom dataset and cells dataset, we can merge them into one by using the join method
#df_tel_cells = pd.merge(df_tel3,df, how='inner', on='Cell_ID')
#df_tel_cells.tail()
#df_tel_cells.info()

df_tel_cells = df_tel3.merge(df_0, left_on='Product', right_on='City')
df_tel_cells #.info()
```
